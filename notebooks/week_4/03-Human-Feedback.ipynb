### Import Dependencies

```python
from langsmith import Client
```

#### Attach Feedback to a specific trace in Langsmith

langsmith trace_id: 019c5e74-e6d1-7a32-bf65-2be6caa46048

```python
client = Client()
```

```python
RUN_ID = "019c5e74-e6d1-7a32-bf65-2be6caa46048"
```

```python
client.create_feedback(
    run_id=RUN_ID,
    key="feedback",
    score=0,
    feedback_source_type="api",
)
```

```python
client.create_feedback(
    run_id=RUN_ID,
    key="comments",
    value="This is a long form free text feedback left by Hanlin",
    feedback_source_type="api",
)
```

```python
client.create_feedback(
    run_id=RUN_ID,
    key="comments_with_score",
    value="This is a long form free text feedback left by Hanlin",
    score=0,
    feedback_source_type="api",
)
```
